# Exercise #1 - Implement a Stack Class Using Only Special Methods

Implement a Stack using only special methods. The following stack functionality needs to be implemented

* `stack + NUM` - Adds the NUM to the stack at the end. Commonly known as `push` in a Stack.
* `stack - 1` - Removes the last element from the stack and returns it. Commonly known as `pop`. Should return the value that was removed.
    * Raise an `IndexError` if this operation is attempted on an empty list
* `print(stack)` - should print out a representation of the stack for developers

CHALLENGE:
* `stack[1]` - Use [`__getitem__()`](https://docs.python.org/3/reference/datamodel.html#object.__getitem__) to be able to look at any element at any index
* `print(len(stack))` - Return the length of the Stack



```python
class Stack:
    # TODO: Implement the special methods this exercise needs:
    #   __init__   - initialize an empty stack
    def __init__(self):
        self.stack = []
    #   __add__    - push onto the stack (used by `stack + N`)
    def __add__(self, num):
        self.stack.append(num)
    #   __sub__    - pop from the stack (used by `stack - 1`); raise IndexError when empty
    def __sub__(self, num):
        if len(self.stack) == 0:
            raise IndexError("Out of bounds")
        return self.stack.pop()
    #   __str__    - developer-facing representation (used by `print(stack)`)
    def __str__(self):
        return f"{self.stack}"
    #   __getitem__- support indexed access (used by `stack[i]`)
    def __getitem__(self, index):
        return self.stack[index]
    #   __len__    - support `len(stack)`
    def __len__(self):
        return len(self.stack)



# Example usage:
stack = Stack()
stack + 1
stack + 2
stack + 3
print(stack)        # Output: [1, 2, 3]
print(len(stack))   # Output: 3
print(stack - 1)    # Output: 3
print(stack - 1)    # Output: 2
print(stack)        # Output: [1]


# Try popping from an empty stack: should raise IndexError
empty = Stack()
try:
    empty - 1
except IndexError as e:
    print(f"Got expected IndexError: {e}")
```
