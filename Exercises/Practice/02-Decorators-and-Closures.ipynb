# Exercise #1 - Write a Debugging Decorator

Write a decorator that prints the name of the function, along with its arguments. The original function should then be executed. Afterwards the decorator should print the result to the screen and return the result.

**Hint**: You can get the name of a function by using `FUNCTION_NAME.__name__`

**CHALLENGE**: Write the decorator to take _any_ number of positional and keyword arguments

````python
def debug(func):
    def wrapper(*args, **kwargs):
        # TODO: print the function name and its arguments
        print(f"Calling {func.__name__} with these args: {args}, {kwargs}")
        result = func(*args, **kwargs)
        # TODO: print the return value, then return it
        print(f"{result=}")
        return result
    return wrapper

# Example usage:
@debug
def add(a, b):
    return a + b

add(3, 5)

# Output
# Calling add with arguments: (3, 5), {}
# Return value: 8
````

# Exercise #2 - Write a Decorator that Always Gives the Answer to the Previous Call

**Hint**: Use closures and `nonlocal`.
