# Exercise #1 - Implement a Reverse-File Context Manager

Implement a Context Manager that reads in a file and reverses the text of the entire file.

Hints/Puzzle Pieces:

1. This method will reverse text across standard out
    ```python
    def reverse(self, text):
        self.stdout(text[::-1])
    ```
2.  This code saves the original `stdout` to a variable and then replaces it with a method you define. Whatever this method does will be used for `stdout`
    ```python    
    import sys
    self.stdout = sys.stdout.write
    sys.stdout.write = A_METHOD_YOU_DEFINE
    ```
3. Be sure to set the original `stdout` back to what it was, otherwise the rest of your output outside the context manager will also be in reverse
4. Run the code below to create a file for you to reverse

````python
pokemon_theme = """I wanna be the very best
Like no one ever was
To catch them is my real test
To train them is my cause
I will travel across the land
Searching far and wide
Teach Pokémon to understand
The power that's inside"""
with open("pokemon_theme.txt", "w") as fh:
    fh.write(pokemon_theme)
````

````python
class ReverseFileRead:
    def __init__(self, path, mode):
        self.path = path
        self.mode = mode
        self.file_handle = None
        
    def __enter__(self):
        import sys
        self.stdout = sys.stdout.write
        sys.stdout.write = self.reverse
        self.file_handle = open(self.path, self.mode)
        return self.file_handle

    def __exit__(self, exc_type, exc_value, traceback):
        import sys
        sys.stdout.write = self.stdout
        self.file_handle.close()
    
    def reverse(self, text):
        self.stdout(text[::-1])


with ReverseFileRead("pokemon_theme.txt", "r") as fh:
    for line in fh.readlines():
        print(line)

print("\nI should be normal")
````
