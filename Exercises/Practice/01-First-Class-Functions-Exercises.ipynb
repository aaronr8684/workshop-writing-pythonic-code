# Exercise #1 - Implement a Higher-Order Function

Higher-Order Functions are functions that either take functions as parameters or return functions as a result of the execution of a function.

* Write a Python function called `apply_operation` that takes three arguments:
  * `operation` - The function to be performed
  * `var1` - The first variable for the operation
  * `var2` - The second variable for the operation
* This function should execute the `operation` function that is passed to it, using the variables as input to the `operation` function.


```python
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

# TODO: Implement the `apply_operation` function
def apply_operation(func, a, b):
    return func(a, b)

print(apply_operation(add, 3, 5))  # Output: 8
print(apply_operation(multiply, 3, 5))  # Output: 15
```

# Exercise #2 - Filtering with Lambda Functions

The [`filter`](https://www.w3schools.com/python/ref_func_filter.asp) function filters the given sequence with the help of a function that tests each element in the sequence to be true or not.

* Write the `lambda` function that, when provided to the `filter` function, will only result in numbers less than or equal to 10 being in the resulting list.

```python
numbers = [2, 15, 8, 12, 6, 10, 18]

# Remove the ADD_LAMBDA_HERE and replace with your own lambda function
filtered_numbers = list(filter(lambda x: x<=10, numbers))
print(filtered_numbers)  # Output: [2, 8, 6, 10]
```

# Exercise #3 - Sorting with Custom Key Functions

Write the code that sorts a list of dictionaries based on the `age` key in each dictionary. Implement this using the `sorted()` function and a lambda function as the key argument.

```python
students = [
    {"name": "Alice", "age": 20},
    {"name": "Bob", "age": 25},
    {"name": "Charlie", "age": 22}
]

# TODO: Sort students by age (ascending) and assign to sorted_students
sorted_students = sorted(students, key=lambda d: d["age"])

print(sorted_students)
# Output: [{'name': 'Alice', 'age': 20}, {'name': 'Charlie', 'age': 22}, {'name': 'Bob', 'age': 25}]
```
